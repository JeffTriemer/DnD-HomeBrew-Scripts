```python
import time
import pandas as pd
import random as rn
global NPCs
global Races
global Pronouns
global Sexualities
global Ages
global Strs
global Dexes
global Cons
global Wises
global Ints
global Chas
global Names
global AgeDescs
AgeDescs =[]
Names =[]
Strs = []
Dexes=[]
Cons=[]
Wises=[]
Ints=[]
Chas=[]
Ages= []
Races = []
Pronouns = []
Sexualities = []
        

def d(number, dice):
    total = 0
    for i in range(number):
        int(rn.randint(1,dice))
        total += int(rn.randint(1,dice))
    return total

def rollStats():
    global Str
    global Dex
    global Con
    global Wis
    global Int
    global Cha
    
    Str = d(3,6)
    Dex = d(3,6)
    Con = d(3,6)
    Wis = d(3,6)
    Int = d(3,6)
    Cha = d(3,6)

def NPC(number):    
    for each in range(number):
        
        global CharacterString
        NameData = pd.read_csv(r"C:\Users\622379\Documents\Python\Homebrew\Names.csv")

        MasculineNames = NameData[NameData['Gender'] == 'Masculine']
        MasculineNames = MasculineNames['Name'].to_list()
        
        FeminineNames = NameData[NameData['Gender'] == 'Feminine']
        FeminineNames = FeminineNames['Name'].to_list()

        races = 'Dwarf', 'Hill Dwarf', 'Mountain Dwarf', 'Elf', 'High Elf', 'Wood Elf','Drow Elf','Halfling','Lightfoot Halfling', 'Stout Halfling', 'Human', 'Dragonborn', 'Gnome','Forest Gnome','Rock Gnome','Half-Elf','Half-Orc','Tiefling'
        ProNouns = 'He/Him', 'She/Her', 'They/Them', 'She/They', 'He/They'
        sexualPreference = 'Masculine', 'Feminine', 'Either'
        # print(len(NameData.index))
        
        raceint = rn.randint(0,17)
        Race = races[raceint]
        Races.append(Race)

        proNounInt = rn.randint(0,4)
        ProNoun = ProNouns[proNounInt]
        Pronouns.append(ProNoun)

        sexualityInt = rn.randint(0,2)
        Sexuality = sexualPreference[sexualityInt]
        
        Sexualities.append(Sexuality)

        rollStats()
        Strs.append(Str)
        Cons.append(Con)
        Dexes.append(Dex)
        Wises.append(Wis)
        Ints.append(Int)
        Chas.append(Cha)

        if 'T' in ProNoun:
            nameInt =rn.randint(0,1542)
            Name = NameData.Name[nameInt]    
        elif "He/Him" == ProNoun:
            nameInt =rn.randint(0,940)
            Name = MasculineNames[nameInt]
        elif "She/Her" == ProNoun:
            nameInt =rn.randint(0,600)
            Name = FeminineNames[nameInt]
        else:
            Name = ""
        Names.append(Name)

        if 'dwarf' in Race.casefold():
            age = rn.randint(0,400)
            Ages.append(age)
            # print(age)
            if age < 51:
                ageDesc = "Child"
            elif age < 150:
                ageDesc = "Young Adult"
            elif age < 250:
                ageDesc = "Middle Aged"
            elif age < 350:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if 'half-orc' in Race.casefold():
            age = rn.randint(0,80)
            Ages.append(age)
            # print(age)
            if age < 14:
                ageDesc = "Child"
            elif age < 25:
                ageDesc = "Young Adult"
            elif age < 55:
                ageDesc = "Middle Aged"
            elif age < 75:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if 'half-elf' in Race.casefold():
            age = rn.randint(0,200)
            Ages.append(age)
            # print(age)
            if age < 20:
                ageDesc = "Child"
            elif age < 75:
                ageDesc = "Young Adult"
            elif age < 120:
                ageDesc = "Middle Aged"
            elif age < 180:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if "elf" in Race.casefold():
            age = rn.randint(0,800)
            Ages.append(age)
            # print(age)
            if age < 101:
                ageDesc = "Child"
            elif age < 300:
                ageDesc = "Young Adult"
            elif age < 500:
                ageDesc = "Middle Aged"
            elif age < 700:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if "tiefling" in Race.casefold():
            age = rn.randint(0,125)
            Ages.append(age)
            # print(age)
            if age < 18:
                ageDesc = "Child"
            elif age < 35:
                ageDesc = "Young Adult"
            elif age < 65:
                ageDesc = "Middle Aged"
            elif age < 100:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if "gnome" in Race.casefold():
            age = rn.randint(0,500)
            Ages.append(age)
            # print(age)
            if age < 40:
                ageDesc = "Child"
            elif age < 200:
                ageDesc = "Young Adult"
            elif age < 350:
                ageDesc = "Middle Aged"
            elif age < 450:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if "dragonborn" in Race.casefold():
            age = rn.randint(0,90)
            Ages.append(age)
            # print(age)
            if age < 15:
                ageDesc = "Child"
            elif age < 30:
                ageDesc = "Young Adult"
            elif age < 60:
                ageDesc = "Middle Aged"
            elif age < 80:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if 'halfling' in Race.casefold():
            age = rn.randint(0,160)
            Ages.append(age)
            # print(age)
            if age < 20:
                ageDesc = "Child"
            elif age < 60:
                ageDesc = "Young Adult"
            elif age < 100:
                ageDesc = "Middle Aged"
            elif age < 145:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"

        if 'human' in Race.casefold():
            age = rn.randint(0,105)
            Ages.append(age)
            # print(age)
            if age < 18:
                ageDesc = "Child"
            elif age < 30:
                ageDesc = "Young Adult"
            elif age < 60:
                ageDesc = "Middle Aged"
            elif age < 90:
                ageDesc = "Old"
            else:
                ageDesc = "Extremely Old"
        
        AgeDescs.append(ageDesc)
        
        statTable = pd.DataFrame({'Str' : [Str], 'Dex' : [Dex], 'Con' : [Con], 'Wis':[Wis], 'Int':[Int], 'Cha':[Cha]})
        CharacterString = f'Name: {Name}\nProNouns: {ProNoun}\nRace: {Race}\nSexual Preference: {Sexuality}\nAge: {ageDesc}, {age}\n{statTable}\n_________________________________'
        
        # print(f'\nName: {Name}')
        # print(f'ProNouns: {ProNoun}')
        # print(f'Race: {Race}')
        # print(f'Sexual Preference: {Sexuality}')
        # print(f'Age: {ageDesc}, {age}\n')
        # print(f'{statTable}')
        # print('____________________________________________________________________________')
        print(CharacterString)
```
